```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
liver_data['Target'] = liver_data['Target'].map({2:0,1:1})
```

```python
liver_data['Target'].value_counts()
```

```python
liver_data['Gender'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
liver_data['Gender'] = le.fit_transform(liver_data['Gender'])
```

1-> Male
0-> Female

```python
liver_data['Gender'].value_counts()
```

```python
x=liver_data.drop(columns=['Target'])
y=liver_data['Target']
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
x.shape,x_train.shape,x_test.shape
```

```python
x_train['Gender'].value_counts(), x_test['Gender'].value_counts()
```

```python
y_train.value_counts(), y_test.value_counts()
```

```python
model=LogisticRegression()
```

```python
scaler=StandardScaler()
```

```python
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.transform(x_test)
```

```python
model.fit(x_train_scaled,y_train)
```

```python
x_train_scaled_prediction=model.predict(x_train_scaled)
training_data_accuracy=accuracy_score(x_train_scaled_prediction,y_train)
training_data_accuracy
```

```python
x_test_scaled_prediction=model.predict(x_test_scaled)
testing_data_accuracy=accuracy_score(x_test_scaled_prediction,y_test)
testing_data_accuracy
```

```python
input_data=(56, 1, 7.65, 5.33, 234, 58, 76, 7.34, 4.213, 0.967)
input_data_as_numpy_array=np.asarray(input_data)
input_data_reshaped=input_data_as_numpy_array.reshape(1,-1)
std_data=scaler.transform(input_data_reshaped)
prediction=model.predict(std_data)
prediction
if (prediction[0]==0):
  print('The person does not have liver cancer')
else:
  print('The person has liver cancer')
```
