```python
x=dementia_data.drop(['Group'],axis=1)
y=dementia_data['Group']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
x.shape, x_train.shape, x_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
x_train_scaled=sc.fit_transform(x_train)
x_test_scaled=sc.transform(x_test)
```

```python
model=LinearRegression()
```

```python
model.fit(x_train_scaled,y_train)
```

```python
x_train_prediction=model.predict(x_train_scaled)
training_data_accuracy=accuracy_score(y_train,x_train_prediction.round())
training_data_accuracy
```

```python
x_test_prediction=model.predict(x_test_scaled)
testing_data_accuracy=accuracy_score(y_test,x_test_prediction.round())
testing_data_accuracy
```

```python
input_data=(2, 764, 1, 0, 83, 20, 4, 28, 0.87, 1772, 0.746, 1.323)
input_data_as_numpy_array=np.asarray(input_data)
input_data_reshaped=input_data_as_numpy_array.reshape(1,-1)
std_data=sc.transform(input_data_reshaped)
prediction=model.predict(std_data)
prediction
if prediction[0]==0:
  print('The person later developed dementia which was initially not the case')
elif prediction[0]==1:
  print('The person has dementia')
else:
  print('The person does not have dementia')
```
