```python
kidney_data['packed_cell_volume'] = pd.to_numeric(kidney_data['packed_cell_volume'], errors='coerce')
kidney_data['white_blood_cell_count'] = pd.to_numeric(kidney_data['white_blood_cell_count'], errors='coerce')
kidney_data['red_blood_cell_count'] = pd.to_numeric(kidney_data['red_blood_cell_count'], errors='coerce')
```

```python
#label encoding for the object data type columns
le = LabelEncoder()
```

```python
kidney_data['red_blood_cells'] = le.fit_transform(kidney_data['red_blood_cells'])
kidney_data['pus_cell'] = le.fit_transform(kidney_data['pus_cell'])
kidney_data['pus_cell_clumps'] = le.fit_transform(kidney_data['pus_cell_clumps'])
kidney_data['bacteria'] = le.fit_transform(kidney_data['bacteria'])
kidney_data['hypertension'] = le.fit_transform(kidney_data['hypertension'])
kidney_data['diabetes_mellitus'] = le.fit_transform(kidney_data['diabetes_mellitus'])
kidney_data['coronary_artery_disease'] = le.fit_transform(kidney_data['coronary_artery_disease'])
kidney_data['appetite'] = le.fit_transform(kidney_data['appetite'])
kidney_data['peda_edema'] = le.fit_transform(kidney_data['peda_edema'])
kidney_data['anemia'] = le.fit_transform(kidney_data['anemia'])
kidney_data['target'] = le.fit_transform(kidney_data['target'])
```

```python
kidney_data
```

```python
x = kidney_data.drop('target', axis=1)
y = kidney_data['target']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
x.shape, y.shape, x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
model = LogisticRegression()
```

```python
scaler = StandardScaler()
```

```python
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
model.fit(x_train, y_train)
```

```python
x_train_prediction = model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction, y_train)
training_data_accuracy
```

```python
x_test_prediction = model.predict(x_test)
test_data_accuracy = accuracy_score(x_test_prediction, y_test)
test_data_accuracy
```

```python
input_data = (48,80,1.02,1,0,1,1,0,121,36,1.2,15,4.4,0.9,13,44,7800,5.2,1,1,0,1,0,0)
input_data_as_numpy_array = np.asarray(input_data)
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)
std_data = scaler.transform(input_data_reshaped)
prediction = model.predict(std_data)
print(prediction)
if (prediction[0] == 1):
  print('The patient does not have chronic kidney disease')
else:
  print('The patient has chronic kidney disease')
```
